```python
import gym
import multi_agents_mno
import numpy as np
```

```python
env_v1 = gym.make('env_3A_3S_9U-v1')
```

```python
action0a = np.array([[[1,0,0],[1,0,0],[1,0,0]]])
action1a = np.array([[[0,1,0],[0,1,0],[0,1,0]]])
action2a = np.array([[[0,0,1],[0,0,1],[0,0,1]]])
obs_n, r_n, _, _ = env_v1.step([action0a, action1a, action2a])
print(r_n)

env_v1.render('env_v1_initial.png')
```

```python
action0b = np.array([[[0,0,1],[1,0,0],[0,1,0]]])
action1b = np.array([[[0,0,1],[1,0,0],[0,1,0]]])
action2b = np.array([[[0,0,1],[1,0,0],[0,1,0]]])
obs_n, r_n, _, _ = env_v1.step([action0b, action1b, action2b])
print(r_n)

env_v1.render('env_v1_worst.png')
```

```python
action0c = np.array([[[0,1,0],[1,0,0],[1,0,0]]])
action1c = np.array([[[0,1,0],[0,0,1],[0,1,0]]])
action2c = np.array([[[0,0,1],[0,0,1],[1,0,0]]])
obs_n, r_n, _, _ = env_v1.step([action0c, action1c, action2c])
print(r_n)

env_v1.render('env_v1_optimal.png')
```
